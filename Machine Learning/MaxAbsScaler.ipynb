```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame([[0.3, 35], [1.2,25],
                   [-0.5, 30], [-1.3,15],
                   [1.1, 19], [0.8, 20]])

df.columns = ['sight', 'age']
df_scaled = df/df.abs().max()

plt.title('Before scaling')
plt.plot(df)
plt.legend(df.columns)
plt.show()

plt.title('After Scaling')
plt.plot(df_scaled)
plt.legend(df_scaled.columns)
plt.show()
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaler.fit(df)
df_new1 = scaler.transform(df)
plt.plot(df_new1)
plt.show()
```
