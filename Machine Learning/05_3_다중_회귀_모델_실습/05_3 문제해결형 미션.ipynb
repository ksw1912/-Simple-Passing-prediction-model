### 미션 : 3차 방정식을 만들고 다항 회귀 모델을 이용하여 예측하기

```python
# 2개의 변곡점을 가진 데이터를 생성 및 시각화
import numpy as np
import matplotlib.pyplot as plt

m = 100
X = 8 * np.random.rand(m, 1) - 4
y = 0.5 * X ** 3 + 0.5 * X ** 2 + X + 3 + np.random.randn(m, 1)
plt.figure(figsize=(6,5))
plt.plot(X, y, "b^")
```

```python
# 다항 회귀 모델을 만들고 점수와 계수, 절편을 출력하는 코드
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_features = PolynomialFeatures(degree = 3,
                                   include_bias = False)
```

### 모델을 만들고 계수, 절편 및 관계 점수를 출력하는 코드 작성해보기

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_poly = poly_features.fit_transform(X)
regr = LinearRegression()
regr.fit(X_poly, y)
print('다항 회귀 모델의 점수 =', regr.score(X_poly, y).round(3))
print('다항 회귀 모델의 계수 =', regr.coef_, '절편 =', regr.intercept_)
```

### 다항 회귀 곡선을 생성하는 기능을 구현하기

```python
X_new = np.linspace(-4, 4, m).reshape(100, 1)
# 이어서 코드 작성

Y_new = poly_features.fit_transform(X_new)
y_predict = regr.predict(Y_new)

plt.figure(figsize=(6,6))
plt.plot(X,y,'b^')
plt.plot(X_new, y_predict, "r-")
```
